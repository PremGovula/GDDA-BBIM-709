```python
# Convert 'InvoiceDate' to datetime if it's not already
df["InvoiceDate"] = pd.to_datetime(df["InvoiceDate"])

# Group by day and aggregate TotalSales and Quantity
daily_agg = df.groupby(df["InvoiceDate"].dt.date).agg({
    "TotalSales": "sum",
    "Quantity": "sum"
}).reset_index()
daily_agg.columns = ["Date", "TotalSales", "Quantity", "InvoiceNo", "StockCode", "Description", "InvoiceDate", "InvoiceDate", "UnitPrice", " CustomerID", "Country", "CustomerName", "Anomaly"]
```
